```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np

# Load the Deep Averaging Network (DAN) version of the Universal Sentence Encoder
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/3")  # Note: This URL corresponds to the DAN version

# Extract sentences and labels
sentences = train_df['text'].astype(str).tolist()  # texts_by_topic_exp
labels = train_df['label'].astype(str).tolist()

# Generate embeddings
embeddings = embed(sentences)

# Access the embeddings from the 'outputs' key
embeddings = embeddings['outputs']

# Convert embeddings and labels to lists of tuples
data_tuples = list(zip(embeddings.numpy(), labels))

# Convert labels to numpy array (optional, depending on your further processing needs)
labels = np.array(labels)
```

### Create centroid

```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np

tensors_list_beginner =  [t[0] for t in data_tuples if t[1] == '0']
tensors_list_expert =  [t[0] for t in data_tuples if t[1] == '1']

# Calculate mean using TensorFlow
mean_tensor_beginner = tf.reduce_mean(tf.stack(tensors_list_beginner), axis=0)
mean_tensor_expert = tf.reduce_mean(tf.stack(tensors_list_expert), axis=0)

# If you need NumPy arrays:
mean_array_beginner = mean_tensor_beginner.numpy()
mean_array_expert = mean_tensor_expert.numpy()
```

```python
import tensorflow as tf
import tensorflow_hub as hub

# Load the Universal Sentence Encoder
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/3")


# Extract sentences and labels
sentences = test_df_2['text']

# Generate embeddings
test_embeddings = embed(sentences)

test_data_tuples = list(zip(test_embeddings, test_df_2['label']))
```

### Measure similatiry btween test embedding and centroids of each label - the closet will determine the label

```python
from sklearn.metrics.pairwise import cosine_similarity
correct = 0
correct_beginner = 0
correct_expert = 0

# Access the 'outputs' key to get the tensor from the dictionary
for embedding, label in zip(test_embeddings['outputs'], test_df_2['label']):
  if label == 0 :
    truth = "beginner"
  else :
    truth = "expert"

  cls_vector1_normalized = embedding.numpy() / np.linalg.norm(embedding.numpy())
  cls_vector2_normalized = mean_tensor_beginner.numpy() / np.linalg.norm(mean_tensor_beginner.numpy())

  similarity_beg = cosine_similarity([cls_vector1_normalized], [cls_vector2_normalized])[0][0]

  cls_vector1_normalized = embedding.numpy() / np.linalg.norm(embedding.numpy())
  cls_vector2_normalized = mean_tensor_expert.numpy() / np.linalg.norm(mean_tensor_expert.numpy())

  similarity_expert = cosine_similarity([cls_vector1_normalized], [cls_vector2_normalized])[0][0]

  if similarity_beg > similarity_expert and truth == "beginner" :
    correct_beginner += 1
    correct += 1
  elif similarity_beg < similarity_expert and truth == "expert" :
    correct_expert += 1
    correct += 1

print("correct : ", correct/len(test_df_2),"\n")
print("correct beginner: ", correct_beginner/len(test_df_2[test_df_2['label'] == 0]),"\n")
print("correct expert: ", correct_expert/len(test_df_2[test_df_2['label'] == 1]),"\n")
```
